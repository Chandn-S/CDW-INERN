```python
from transformers import pipeline

# Load a pretrained emotion analysis pipeline
emotion_analyzer = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
```

```python
# Input sentences
text = ["I am so happy today!", "I feel really sad and lost."]

# Predict emotions
predictions = emotion_analyzer(text)

# Print results
for sentence, pred in zip(text, predictions):
    print(f"Text: {sentence}")
    print(f"Predicted Emotion: {pred['label']} (Confidence: {pred['score']:.2f})\n")
```
